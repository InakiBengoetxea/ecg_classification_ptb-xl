## Overview
This notebook executes a comprehensive evaluation suite to stress-test the diagnostic stability of the fully trained multi-modal ECG model under simulated clinical noise. The pipeline measures performance degradation without altering the underlying network knowledge.

## Key Objectives:
* **Frozen Evaluation:** Load the finalized, optimized weights from `best_fusion_model.pt` and enforce strict evaluation mode (`model.eval()`).
* **Targeted Perturbation:** Isolate the hidden test dataset (**Fold 10**) as the exclusive target for synthetic artifact injection, leaving Folds 1–9 entirely untouched.
* **Degradation Mapping:** Sweep across systematically decreasing Signal-to-Noise Ratio (SNR) levels (Infinity dB, 24 dB, 18 dB, 12 dB, 6 dB, and 0 dB) to construct a precise performance degradation trajectory using Macro AUC scores.

```python
# ==============================================================================
# AUTOMATED MULTI-TASK FUSION MODEL ROBUSTNESS EVALUATION PIPELINE
# ==============================================================================
import os
import sys
import torch
import numpy as np
import pandas as pd
import importlib
import matplotlib.pyplot as plt
from pathlib import Path

# Step 1: Append workspace path dynamically to resolve local utility/model imports
PROJECT_ROOT = Path(os.getcwd()).parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

# Import required sub-modules and utilities
import utils.data_loader
importlib.reload(utils.data_loader)
from utils.data_loader import load_ptbxl, load_metadata

from utils.robustness import run_comprehensive_stress_test  # type: ignore
from models.meta_mlp import MetaMLP
from models.xresnet1d import xresnet1d101
from models.fusion import FusionModel

# Global configurations matching training parameters
PATH = Path("../ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3/")
SAMPLING_RATE = 100
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
test_fold = 10  # Target split for evaluations

if not PATH.exists():
    print(f"ERROR: Dataset directory not found at {PATH.absolute()}")

# Automated loop across all target classification horizons
TASKS_TO_RUN = ['superclass', 'subclass', 'scp-statement']
all_task_results = {}

for current_task in TASKS_TO_RUN:
    print("\n" + "="*80)
    print(f"STARTING ROBUSTNESS EVALUATION FOR TARGET COMPONENT: [{current_task.upper()}]")
    print("="*80)
    
    # 1. Dynamically load evaluation signals and labels for the active task
    _, _, X_test, _, _, y_test, classes, Y_filtered = load_ptbxl(
        PATH, SAMPLING_RATE, task=current_task
    )
    num_classes = len(classes)
    print(f" -> ECG Signals Matrix Shape:  {X_test.shape}")
    print(f" -> Class Targets Count:        {num_classes}")

    # 2. Extract and align matching test metadata samples using task filters
    meta_test = load_metadata(Y_filtered[Y_filtered.strat_fold == test_fold]).reset_index(drop=True)
    
    # Standardize input tracking columns matching training steps
    num_cols = ["age", "sex", "height", "weight", "pacemaker", "extra_beats", "heart_axis"]
    meta_test[num_cols] = meta_test[num_cols].astype(float)
    X_meta_test = meta_test[num_cols].values
    meta_features = X_meta_test.shape[1]
    print(f" -> Aligned Metadata Shape:    {X_meta_test.shape}")

    # 3. Reconstruct multi-modal network architectures dynamically
    ecg_branch = xresnet1d101(input_channels=12, num_classes=num_classes).to(DEVICE)
    meta_branch = MetaMLP(in_features=meta_features).to(DEVICE)
    
    fusion_model = FusionModel(
        ecg_model=ecg_branch, 
        meta_model=meta_branch, 
        num_classes=num_classes
    ).to(DEVICE)

    # 4. Load the trained fusion checkpoint saved by the training loop
    checkpoint_path = f"../outputs/best_fusion_model_{current_task}.pt"
    try:
        fusion_model.load_state_dict(torch.load(checkpoint_path, map_location=DEVICE))
        print(f" -> Successfully loaded target weights: {checkpoint_path}")
    except FileNotFoundError:
        print(f" -> WARNING: Checkpoint not found at {checkpoint_path}. Attempting alternative path lookup...")
        checkpoint_path = PROJECT_ROOT / "outputs" / current_task / "best_fusion_model.pt"
        fusion_model.load_state_dict(torch.load(checkpoint_path, map_location=DEVICE))
        print(f" -> Successfully loaded model weights from fallback: {checkpoint_path}")
        
    fusion_model.eval()

    # 5. Continuous Stream Wrapper to synchronize tabular features with evaluation loops
    class FusionStressTestWrapper(torch.nn.Module):
        def __init__(self, model_container, metadata_array, execution_device):
            super().__init__()
            self.fusion_model = model_container
            self.meta_tensor = torch.tensor(metadata_array, dtype=torch.float32).to(execution_device)
            self.current_idx = 0
            
        def forward(self, x_ecg):
            batch_size = x_ecg.size(0)
            
            # Guard conditions: automatically reset index track boundaries across noise groups
            if self.current_idx >= len(self.meta_tensor):
                self.current_idx = 0
            if self.current_idx + batch_size > len(self.meta_tensor):
                self.current_idx = 0
                
            x_meta = self.meta_tensor[self.current_idx : self.current_idx + batch_size]
            self.current_idx += batch_size
            return self.fusion_model(x_ecg, x_meta)

    wrapped_model = FusionStressTestWrapper(fusion_model, X_meta_test, DEVICE)

    # 6. Execute structured stress perturbations over the active model layout
    df_results = run_comprehensive_stress_test(wrapped_model, X_test, y_test, device=DEVICE)
    all_task_results[current_task] = df_results

    # 7. Generate and Export Task-Specific Visualizations
    plt.figure(figsize=(9, 5.5))
    
    # Render baseline reference line
    clean_val = df_results[df_results['Noise Type'] == 'Clean']['Macro_AUC'].values[0]
    plt.axhline(y=clean_val, color='black', linestyle='--', alpha=0.7, label=f'Clean Baseline ({clean_val:.3f})')
    
    # Configure plotting parameters
    snrs = [24, 18, 12, 6, 0]
    markers = {'Baseline Wander': '-o', 'Muscle Artifact (EMG)': '-s', 'Power Line (50Hz)': '-^'}
    colors = {'Baseline Wander': '#E66101', 'Muscle Artifact (EMG)': '#5E3C99', 'Power Line (50Hz)': '#2B83BA'}
    
    for noise in markers.keys():
        subset = df_results[df_results['Noise Type'] == noise].sort_values(by='SNR_dB', ascending=False)
        plt.plot(snrs, subset['Macro_AUC'], markers[noise], color=colors[noise], label=noise, linewidth=2, markersize=7)
        
    plt.title(f"Fusion Performance Degradation - Task: [{current_task.upper()}]", fontsize=12, fontweight='bold')
    
    # Added raw string prefix 'r' to prevent carriage return rendering
    plt.xlabel(r"Signal-to-Noise Ratio (SNR in dB) $\rightarrow$ Increasing Noise Severity", fontsize=11)
    
    plt.ylabel("Performance Metric (Macro AUC)", fontsize=11)
    plt.gca().invert_xaxis()  # Places higher-noise configurations on the right boundary
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend(loc='lower left', frameon=True)
    plt.ylim(0.45, 1.0)
    
    # Save chart layout to corresponding outputs target directory
    save_dir = Path("../outputs")
    save_dir.mkdir(parents=True, exist_ok=True)
    output_path = save_dir / f"robustness_degradation_{current_task}.png"
    
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    print(f"[SAVED] Degradation analysis graphic generated at: {output_path.resolve()}")
    plt.show() 

print("\n" + "="*80)
print("ALL MULTI-TASK ROBUSTNESS EVALUATIONS COMPLETED SUCCESSFULLY.")
print("="*80)
```
