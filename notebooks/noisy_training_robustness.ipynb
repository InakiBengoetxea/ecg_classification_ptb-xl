## Overview
This notebook implements a targeted data augmentation pipeline to evaluate model robustness under controlled training corruption. Exactly **10% of the training data (Folds 1–8)** is randomly selected and injected with clinical noise artifacts at a fixed intensity . 

## Key Objectives:
* **Data Augmentation:** Evaluate if exposing the network to minor artifacts during the learning phase acts as a regularizer.
* **Pure Evaluation:** Validate the training progress against a completely clean validation set (**Fold 9**).
* **Generalization Test:** Test the finalized model against a 100% clean test set (**Fold 10**) to determine if training-phase noise injection preserves or enhances baseline diagnostic accuracy.

```python
# ==============================================================================
# AUTOMATED ISOLATED NOISY TRAINING & STATISTICAL SIGNIFICANCE PIPELINE
# ==============================================================================
import os
import sys
import torch
import numpy as np
import pandas as pd
import importlib
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import roc_auc_score
from scipy.stats import ttest_rel

# Step 1: Append workspace path dynamically to resolve local utility/model imports
PROJECT_ROOT = Path(os.getcwd()).parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

# Import required sub-modules and utilities
import utils.data_loader
importlib.reload(utils.data_loader)
from utils.data_loader import load_ptbxl, load_metadata

# Import noise injection primitives for training augmentation
from utils.robustness import inject_baseline_wander, inject_muscle_artifact, inject_power_line_interference
from models.meta_mlp import MetaMLP
from models.xresnet1d import xresnet1d101
from models.fusion import FusionModel

# Global configurations matching standard network parameters
PATH = Path("../ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3/")
SAMPLING_RATE = 100
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
BATCH_SIZE = 64
EPOCHS = 15       # Increased to allow early stopping to function meaningfully
PATIENCE = 3      # Number of epochs to wait for validation improvement before halting
TARGET_TRAINING_SNR = 12  
N_BOOTSTRAP = 30  # Number of bootstrap resamples for t-test validity

if not PATH.exists():
    print(f"ERROR: Dataset directory not found at {PATH.absolute()}")

# Automated loop across all target classification horizons
TASKS_TO_RUN = ['superclass', 'subclass', 'scp-statement']
statistical_master_records = []

for current_task in TASKS_TO_RUN:
    print("\n" + "="*90)
    print(f"STARTING SEQUENTIAL AUGMENTATION & STATISTICAL ANALYSIS FOR TASK: [{current_task.upper()}]")
    print("="*90)
    
    # 1. Load complete matrix splits for the active classification target
    X_train, X_val, X_test, y_train, y_val, y_test, classes, Y_filtered = load_ptbxl(
        PATH, SAMPLING_RATE, task=current_task
    )
    num_classes = len(classes)
    
    # 2. Extract and align structured tabular metadata fields across all splits
    num_cols = ["age", "sex", "height", "weight", "pacemaker", "extra_beats", "heart_axis"]
    meta_train = load_metadata(Y_filtered[Y_filtered.strat_fold <= 8]).reset_index(drop=True)
    meta_val = load_metadata(Y_filtered[Y_filtered.strat_fold == 9]).reset_index(drop=True)
    meta_test = load_metadata(Y_filtered[Y_filtered.strat_fold == 10]).reset_index(drop=True)
    
    for df in [meta_train, meta_val, meta_test]:
        df[num_cols] = df[num_cols].astype(float)
        
    X_meta_train = meta_train[num_cols].values
    X_meta_val = meta_val[num_cols].values
    X_meta_test = meta_test[num_cols].values
    meta_features = X_meta_train.shape[1]

    # Convert static evaluation datasets upfront to clean tensors
    X_val_tensor = torch.tensor(X_val, dtype=torch.float32).permute(0, 2, 1)
    X_meta_val_tensor = torch.tensor(X_meta_val, dtype=torch.float32)
    
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32).permute(0, 2, 1)
    X_meta_test_tensor = torch.tensor(X_meta_test, dtype=torch.float32)

    # Dictionary to collect prediction matrices for bootstrapping analysis
    model_predictions = {}

    # ==========================================================================
    # STEP 3: EVALUATE BASELINE CLEAN MODEL (LOAD FROM DISK)
    # ==========================================================================
    print(f" -> Loading pre-trained clean baseline model checkpoint...")
    base_ecg = xresnet1d101(input_channels=12, num_classes=num_classes).to(DEVICE)
    base_meta = MetaMLP(in_features=meta_features).to(DEVICE)
    baseline_model = FusionModel(ecg_model=base_ecg, meta_model=base_meta, num_classes=num_classes).to(DEVICE)
    
    checkpoint_path = f"../outputs/best_fusion_model_{current_task}.pt"
    try:
        baseline_model.load_state_dict(torch.load(checkpoint_path, map_location=DEVICE, weights_only=True))
    except FileNotFoundError:
        checkpoint_path = PROJECT_ROOT / "outputs" / current_task / "best_fusion_model.pt"
        baseline_model.load_state_dict(torch.load(checkpoint_path, map_location=DEVICE, weights_only=True))
        
    baseline_model.eval()
    base_preds = []
    with torch.no_grad():
        for i in range(0, X_test_tensor.size(0), BATCH_SIZE):
            logits = baseline_model(X_test_tensor[i : i + BATCH_SIZE].to(DEVICE), X_meta_test_tensor[i : i + BATCH_SIZE].to(DEVICE))
            base_preds.append(torch.sigmoid(logits).cpu().numpy())
    model_predictions['Baseline (Clean)'] = np.vstack(base_preds)

    # ==========================================================================
    # STEP 4: SEQUENTIAL NOISY TRAINING PIPELINE LOOPS WITH EARLY STOPPING
    # ==========================================================================
    noise_strategies = {
        'Baseline Wander Regularization': inject_baseline_wander,
        'Muscle Artifact Regularization': inject_muscle_artifact,
        'Power Line Regularization': inject_power_line_interference
    }

    for label, noise_function in noise_strategies.items():
        print(f"\n --> Launching Isolated Routine: [{label.upper()}]")
        
        # Isolate exactly 10% unique random training vectors
        rng = np.random.default_rng(seed=42)
        num_train_samples = X_train.shape[0]
        num_noisy_samples = int(0.10 * num_train_samples)
        noisy_indices = rng.choice(num_train_samples, size=num_noisy_samples, replace=False)
        
        X_train_augmented = X_train.copy()
        X_train_augmented[noisy_indices] = noise_function(X_train_augmented[noisy_indices], snr_db=TARGET_TRAINING_SNR)
        
        # Prepare training tensors
        X_train_tensor = torch.tensor(X_train_augmented, dtype=torch.float32).permute(0, 2, 1)
        X_meta_train_tensor = torch.tensor(X_meta_train, dtype=torch.float32)
        y_train_tensor = torch.tensor(y_train, dtype=torch.float32)

        # Initialize unique un-optimized networks
        ecg_branch = xresnet1d101(input_channels=12, num_classes=num_classes).to(DEVICE)
        meta_branch = MetaMLP(in_features=meta_features).to(DEVICE)
        fusion_model = FusionModel(ecg_model=ecg_branch, meta_model=meta_branch, num_classes=num_classes).to(DEVICE)
        
        criterion = torch.nn.BCEWithLogitsLoss()
        optimizer = torch.optim.Adam(fusion_model.parameters(), lr=1e-3)

        # Early stopping metrics tracking initializations
        best_val_auc = 0.0
        patience_counter = 0
        save_name = label.lower().replace(" ", "_")
        target_weight_path = f"../outputs/augmented_model_{save_name}_{current_task}.pt"

        # Training Execution block
        for epoch in range(EPOCHS):
            fusion_model.train()
            permutation = torch.randperm(X_train_tensor.size(0))
            epoch_loss = 0.0
            
            for i in range(0, X_train_tensor.size(0), BATCH_SIZE):
                indices = permutation[i : i + BATCH_SIZE]
                optimizer.zero_grad()
                logits = fusion_model(X_train_tensor[indices].to(DEVICE), X_meta_train_tensor[indices].to(DEVICE))
                loss = criterion(logits, y_train_tensor[indices].to(DEVICE))
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item() * len(indices)
                
            # Internal evaluation check against clean Validation Split
            fusion_model.eval()
            val_preds = []
            with torch.no_grad():
                for i in range(0, X_val_tensor.size(0), BATCH_SIZE):
                    logits = fusion_model(X_val_tensor[i : i + BATCH_SIZE].to(DEVICE), X_meta_val_tensor[i : i + BATCH_SIZE].to(DEVICE))
                    val_preds.append(torch.sigmoid(logits).cpu().numpy())
            val_auc = roc_auc_score(y_val, np.vstack(val_preds), average='macro')
            print(f"    [Epoch {epoch+1:02d}/{EPOCHS:02d}] Loss: {epoch_loss/num_train_samples:.4f} | Clean Val AUC: {val_auc:.4f}")

            # Early Stopping Check
            if val_auc > best_val_auc:
                best_val_auc = val_auc
                patience_counter = 0
                # Preserve the absolute highest-performing weights encountered
                torch.save(fusion_model.state_dict(), target_weight_path)
            else:
                patience_counter += 1
                if patience_counter >= PATIENCE:
                    print(f"    --> Early stopping triggered at epoch {epoch+1:02d}. Reverting to best weights.")
                    break

        # Restore optimal parameters prior to executing the test collection phase
        fusion_model.load_state_dict(torch.load(target_weight_path, map_location=DEVICE, weights_only=True))

        # Gather clean test predictions using optimal checkpoint weights
        fusion_model.eval()
        aug_preds = []
        with torch.no_grad():
            for i in range(0, X_test_tensor.size(0), BATCH_SIZE):
                logits = fusion_model(X_test_tensor[i : i + BATCH_SIZE].to(DEVICE), X_meta_test_tensor[i : i + BATCH_SIZE].to(DEVICE))
                aug_preds.append(torch.sigmoid(logits).cpu().numpy())
        model_predictions[label] = np.vstack(aug_preds)

    # ==========================================================================
    # STEP 5: BOOTSTRAPPED DISTRIBUTION & T-TEST GENERATION
    # ==========================================================================
    print(f"\n -> Launching statistical bootstrapping analysis ({N_BOOTSTRAP} iterations)...")
    bootstrap_distributions = {cfg: [] for cfg in model_predictions.keys()}
    n_test_samples = y_test.shape[0]
    boot_rng = np.random.default_rng(seed=100)

    for _ in range(N_BOOTSTRAP):
        boot_idx = boot_rng.choice(n_test_samples, size=n_test_samples, replace=True)
        y_true_boot = y_test[boot_idx]
        
        # Skip iteration balances if a rare binary class combination drops to zero variance
        try:
            for cfg, preds in model_predictions.items():
                auc_score = roc_auc_score(y_true_boot, preds[boot_idx], average='macro')
                bootstrap_distributions[cfg].append(auc_score)
        except ValueError:
            continue

    # Perform statistical paired t-testing against clean baseline distribution
    baseline_dist = bootstrap_distributions['Baseline (Clean)']
    boxplot_data = [baseline_dist]
    boxplot_labels = ['Baseline']
    
    print(f"\n--- Statistical Significance Results vs. Baseline ---")
    for cfg in noise_strategies.keys():
        current_dist = bootstrap_distributions[cfg]
        boxplot_data.append(current_dist)
        boxplot_labels.append(cfg.split()[0] + " Reg.") # Shortens label text
        
        t_stat, p_val = ttest_rel(baseline_dist, current_dist)
        mean_auc = np.mean(current_dist)
        is_significant = "YES" if p_val < 0.05 else "NO"
        
        print(f" * {cfg:32s} -> Mean AUC: {mean_auc:.4f} | p-value: {p_val:.4e} | Significant change? {is_significant}")
        
        statistical_master_records.append({
            'Task': current_task,
            'Configuration': cfg,
            'Mean_Bootstrap_AUC': mean_auc,
            'P_Value': p_val,
            'Statistically_Significant': is_significant
        })

    # ==========================================================================
    # STEP 6: PLOT PERFORMANCE DISTRIBUTION TRAJECTORIES
    # ==========================================================================
    plt.figure(figsize=(8, 5))
    median_props = dict(linestyle='-', linewidth=2, color='#D95F02')
    box_props = dict(linestyle='-', linewidth=1.5, color='#1B9E77')
    
    plt.boxplot(boxplot_data, labels=boxplot_labels, patch_artist=False, medianprops=median_props, boxprops=box_props)
    plt.title(f"Regularization Shift & Variance Profiles - Task: [{current_task.upper()}]", fontsize=11, fontweight='bold')
    plt.ylabel("Bootstrapped Macro AUC Score", fontsize=11)
    plt.grid(True, linestyle='--', alpha=0.4)
    
    save_dir = Path("../outputs")
    save_dir.mkdir(parents=True, exist_ok=True)
    plt.savefig(save_dir / f"noise_regularization_significance_{current_task}.png", dpi=300, bbox_inches='tight')
    plt.show()

# Final Global Output Display
print("\n" + "="*90)
print("ALL MULTI-TASK ISOLATED TRAINING & PLOTTING TASKS EXECUTED SUCCESSFULLY.")
print("="*90)
df_master_summary = pd.DataFrame(statistical_master_records)
print(df_master_summary.to_string(index=False))
```

A single, finalised, frozen model checkpoint that corresponds to the lowest validation loss is used for all assessments. On the held-out test partition (Fold 10), statistical bootstrapping (N = 30 iterations) is used to create the variance profile.This optimal model checkpoint is applied to the uncorrupted, noise-free test records to establish the Baseline (Clean) data. To construct a pseudo-independent patient cohort, the test dataset is randomly resampled with replacement during each of the 30 bootstrap iterations. The same resampled slice is then used to test both the clean baseline model and the model configuration that is being evaluated. Instead of using a single static scalar threshold, this produces a paired distribution of Macro AUC scores, guaranteeing that the relative p-values take natural patient-sampling variance into account.
